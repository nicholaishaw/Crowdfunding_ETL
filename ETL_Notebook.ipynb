### Create the Category and Subcategory DataFrames
---
To better organize the data and prepare it for SQL upload, we needed to first create a separate 'Category" dataframe and a 'Subcategory' dataframe from our original excel spreadsheet.

**The Category DataFrame will have the following columns:**
- A "category_id" column that is numbered sequential form 1 to the length of the number of unique categories.
- A "category" column that has only the categories.

The Category dataframe will be exported as a `category.csv` CSV file before SQL upload.

**The SubCategory DataFrame will have the following columns:**
- A "subcategory_id" column that is numbered sequential form 1 to the length of the number of unique subcategories.
- A "subcategory" column that has only the subcategories. 

the SubCategory DataFrame will be exported as a `subcategory.csv` CSV file before SQL upload

```python
# Getting the crowdfunding_info_df columns.
crowdfunding_info_df.keys()
```

```python
# Assigning the category and subcategory values to category and subcategory columns.
crowdfunding_info_df[['category', 'subcategory']] = crowdfunding_info_df['category & sub-category'].str.split('/', expand= True)
crowdfunding_info_df
```

```python
# Getting the unique values in the category column
crowdfunding_info_df['category'].unique()
```

```python
# Getting the unique categories and subcategories in separate lists.
categories = crowdfunding_info_df['category'].unique()
subcategories = crowdfunding_info_df['subcategory'].unique()
print(categories)
print(subcategories)
```

```python
# Getting the number of distinct values in the categories and subcategories lists.
print(len(categories))
print(len(subcategories))
```

```python
# Creating numpy arrays from 1-9 for the categories and 1-24 for the subcategories.
category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1, 25)

print(category_ids)
print(subcategory_ids)
```

```python
# Using a list comprehension to add "cat" to each category_id.
# Using a list comprehension to add "subcat" to each subcategory_id.

cat_ids = ["cat" + str(category) for category in category_ids]
scat_ids = ["subcat" + str(subcategory) for subcategory in subcategory_ids]   
    
print(cat_ids)
print(scat_ids)
```

```python
# Creating a category DataFrame with the category_id array as the category_id and categories list as the category name.
category_df = pd.DataFrame()
category_df['category_id'] = cat_ids
category_df['category'] = categories

# Creating a category DataFrame with the subcategory_id array as the subcategory_id and subcategories list as the subcategory name. 
subcategory_df = pd.DataFrame()
subcategory_df['subcategory_id'] = scat_ids
subcategory_df['subcategory'] = subcategories
```

```python
# Previewing the category dataframe
category_df
```

```python
# Previewing the subcategory dataframe
subcategory_df
```

```python
# Exporting categories_df and subcategories_df as CSV files.
category_df.to_csv("Resources/category.csv", index=False)

subcategory_df.to_csv("Resources/subcategory.csv", index=False)
```

```python
# Similarly, you can also get the group the data by category and assign each category a unique value

groupedcategory_df = pd.DataFrame()
groupedcategory_df['category'] = crowdfunding_info_df['category'].values
groupedcategory_df['category_id'] = groupedcategory_df.groupby('category', sort = False).ngroup() + 1
groupedcategory_df
```

```python
# You can also do the same with the subcategories
grouped_subcategory_df = pd.DataFrame()
grouped_subcategory_df['subcategory'] = crowdfunding_info_df['subcategory'].values
grouped_subcategory_df['subcategory_id'] = grouped_subcategory_df.groupby('subcategory', sort = False).ngroup() + 1
grouped_subcategory_df
```

### Campaign DataFrame
----
After the category and subcategory dataframes have been created, I also had to create a master campaign dataframe that contains the information on each crowdfunding activity and organization information.

**Create a Campaign DataFrame that has the following columns:**
- The "cf_id" column.
- The "contact_id" column.
- The “company_name” column.
- The "blurb" column is renamed as "description."
- The "goal" column.
- The "goal" column is converted to a `float` datatype.
- The "pledged" column is converted to a `float` datatype. 
- The "backers_count" column. 
- The "country" column.
- The "currency" column.
- The "launched_at" column is renamed as "launch_date" and converted to a datetime format. 
- The "deadline" column is renamed as "end_date" and converted to a datetime format.
- The "category_id" with the unique number matching the “category_id” from the category DataFrame. 
- The "subcategory_id" with the unique number matching the “subcategory_id” from the subcategory DataFrame.
- And, create a column that contains the unique four-digit contact ID number from the `contact.xlsx` file.
 

The campaign DataFrame will be saved as `campaign.csv` CSV file before SQL upload.


```python
# Creating a copy of the campaign_df
campaign_df = crowdfunding_info_df.copy()
campaign_df.head()
```

```python
# Renaming the blurb, launched_at, and deadline columns.
campaign_df.rename(columns={'blurb': 'description', 'launched_at': 'launched_date', 'deadline' : 'end_date'}, inplace=True)
campaign_df
```

```python
# Converting the goal and pledged columns to a `float` data type.
campaign_df['goal'] = campaign_df['goal'].astype(float)
campaign_df['pledged'] = campaign_df['pledged'].astype(float)
campaign_df
```

```python
# Checking the datatypes
campaign_df.dtypes
```

```python
# Formatting the launched_date and end_date columns to datetime format
from datetime import datetime as dt

campaign_df['launched_date'] = pd.to_datetime(campaign_df['launched_date'])
campaign_df['end_date'] = pd.to_datetime(campaign_df['end_date'])
campaign_df
```

```python
# Merging the campaign_df with the category_df on the "category" column and the subcategory_df on the "subcategory" column.

campaign_merged_df = pd.merge(campaign_df, category_df, on="category")
campaign_merged_df = pd.merge(campaign_merged_df, subcategory_df, on="subcategory")
campaign_merged_df.tail(10)
```

```python
# Dropping unwanted columns
campaign_cleaned = campaign_merged_df.drop(['staff_pick', 'spotlight', 'category & sub-category', 'category', 'subcategory'], axis=1)
campaign_cleaned
```

```python
# Exporting the DataFrame as a CSV file. 
campaign_cleaned.to_csv("Resources/campaign.csv", index=False)
```

### Create the Contacts DataFrame 
---
Finally, I needed to create a 'Contacts' dataframe that will hold the contact information for each organization.

**Create a Contacts DataFrame that has the following columns:**
- A column named "contact_id"  that contains the unique number of the contact person.
- A column named "first_name" that contains the first name of the contact person.
- A column named "last_name" that contains the first name of the contact person.
- A column named "email" that contains the email address of the contact person

The Contacts DataFrame will be saved as a `contacts.csv` CSV file before SQL upload.

### Using Pandas to create the contacts DataFrame.

```python
# Iterating through the contact_info_df and convert each row to a dictionary.
import json
dict_values = []
for i, row in contact_info_df.iterrows():
    contact = row['contact_info']
    clean_contact = json.loads(contact)
    values = [v for k, v in clean_contact.items()]
    dict_values.append(values)

# Print out the list of values for each row.
print(dict_values)
```

```python
# Creating a contact_info DataFrame and add each list of values, i.e., each row 
# to the 'contact_id', 'name', 'email' columns.
contact_df = pd.DataFrame(dict_values, columns=['contact_id', 'name', 'email'])
contact_df
```

```python
# Checking the datatypes.
contact_df.dtypes
```

```python
# Creating a "first"name" and "last_name" column with the first and last names from the "name" column. 
contact_df[['first_name', 'last_name']] = contact_df['name'].str.split(' ', expand= True)

# Dropping the name column
contact_df = contact_df.drop(columns=['name'], axis=1)
contact_df
```

```python
# Reordering the columns
contacts_df_clean = contact_df[['contact_id', 'first_name', 'last_name', 'email']]
contacts_df_clean
```

```python
# Checking the datatypes one more time before exporting as CSV file.
contacts_df_clean.dtypes
```

```python
# Exporting the DataFrame as a CSV file. 
contacts_df_clean.to_csv("Resources/contacts.csv", encoding='utf8', index=False)
```
